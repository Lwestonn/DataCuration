1. There are three species of flowers in the dataset. The sepal length-width combination of one of the species is quite distinct from that of the other two species. Make an appropriate visualization plot to find out the distinct species. (3 points)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'sample_data/iris.data'
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(file_path, header=None, names=column_names)


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', style='species')
plt.title('sepal length vs sepal width per species')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend(title='species')
plt.show()
```

2. Make a plot that shows you the distribution of petal length for each of the three species. It should show median, 25th and 75th percentile values and outliers (2 points).

```python
#for some reason it loaded in as objects which prevented me from being able to plot the values so I needed to change the values to numeric
df['sepal_length'] = pd.to_numeric(df['sepal_length'], errors='coerce')
df['sepal_width'] = pd.to_numeric(df['sepal_width'], errors='coerce')
df['petal_length'] = pd.to_numeric(df['petal_length'], errors='coerce')
df['petal_width'] = pd.to_numeric(df['petal_width'], errors='coerce')

plt.figure(figsize=(10, 6))
#use seaborn
sns.boxplot(data=df, x='species', y='petal_length', palette='Set2')
plt.title('Petal Length Distribution')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

3. Plot a heat map depicting the correlation between each pair of attributes. Show the corresponding correlation values within each cell. Which pair of flower attributes are the most highly correlated? (3 points)
