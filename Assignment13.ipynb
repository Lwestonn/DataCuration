1. Which three genres have the highest danceability on average? (2 point)

```python
import pandas as pd

#create table
path = 'sample_data/spotify_data.csv'
df = pd.read_csv(path)

avgdf = df.groupby('genres')['danceability'].mean()
topThree = avgdf.sort_values(ascending=False).head(3)
topThree
```

2. Use 'track_popularity' to create a new column 'track_popularity_level' by binning track_popularity into 3 bins. Label the bins 'low', 'medium' and 'high' (Hint: Use pd.cut). (2 points)

```python
# Creating a new column 'track_popularity_level' by binning 'track_popularity' into 3 bins
df['track_popularity_level'] = pd.cut(df['track_popularity'], bins=3, labels=['low', 'medium', 'high'])
df
```

3. Create a pivot table to show the average danceability of tracks grouped by genre and track popularity level on the rows and separated by energy levels in the columns. Show the margin averages. What does this pivot table tell you about the relationship between popularity and danceability for a) country tracks and b) electronic tracks? (2 points)

```python
pivot = df.pivot_table(values='danceability', index=['genres', 'track_popularity_level'], columns='energy_lvl', aggfunc='mean', margins=True)
pivot.loc[['country', 'electronic']]
```

4. What percentage of all the unique tracks are contributed by the top 3 artists of each genre, where the top artists are based on artist_popularity, and the unique tracks are based on unique values of track_name? A typical approach that will not work: If you group the data by genre, and filter the top 3 rows by artist_popularity, then you may not get 3 unique artists, as one artist can have multiple tracks.

```python
topGenre = df.groupby(['genres', 'artist_name'])['artist_popularity'].max()
topThreeGenre = topGenre.groupby('genres').nlargest(3).reset_index(level=0, drop=True)

topArtists = df[df.set_index(['genres', 'artist_name']).index.isin(topThreeGenre.index)]
uniqueTracks = topArtists['track_name'].nunique()
totalUnique = df['track_name'].nunique()

finalPercent = (uniqueTracks / totalUnique) * 100
finalPercent
```
