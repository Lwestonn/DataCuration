1. Read the data movies dataset from this Link to an external site. into a Pandas dataframe. How many movies are there in the dataset? (1 point)

```python
import pandas as pd
#dont need other imports such as json or numpy just this

url = 'https://raw.githubusercontent.com/vega/vega-datasets/master/data/movies.json'

df = pd.read_json(url)


movies = len(df)
print("There are", movies,  "movies in the data set")
```

2. Drop all columns except 'Title', 'US Gross', 'Worldwide Gross', ‘US DVD Sales’, 'Production Budget', and ’IMDB Rating'.

```python
keep = ['Title', 'US Gross', 'Worldwide Gross', 'US DVD Sales', 'Production Budget', 'IMDB Rating']

df = df.loc[:, keep]
df
```

3. Set 'Title' as the index of your dataframe.

```python
#set_index
df.set_index('Title', inplace=True)
df
```

4. Assume return on investment (ROI) for a movie is defined as:
𝑅𝑂𝐼 = 𝑊𝑜𝑟𝑙𝑑𝑤𝑖𝑑𝑒 𝑔𝑟𝑜𝑠𝑠 − 𝑃𝑟𝑜𝑑𝑢𝑐𝑡𝑖𝑜𝑛 𝑏𝑢𝑑𝑔𝑒𝑡
What is the mean ROI? Which movie has the maximum ROI? Is it the one with the minimum production budget?

```python
#create roi for each in the actual df
#df['ROI'] = df['Worldwide Gross'] / df['Production Budget']
df['ROI'] = df['Worldwide Gross'] - df['Production Budget']

# calculate roi
meanRoi = df['ROI'].mean()
print(f"Mean ROI: {meanRoi}")

#find max roi
maxRoi = df['ROI'].idxmax()
print(f"Movie with maximum ROI: {maxRoi}")

#check movie with max to see if it has minimum budget
minBudget = df['Production Budget'].idxmin()
#simple if else
if maxRoi == minBudget:

    print(f"The movie with the max ROI, '{maxRoi}', has the minimum budget as well.")

else:

    print(f"The movie with the max ROI, '{maxRoi}', does not have the minimum production budget. '{minBudget}' is the min.")
```

5. What is the ROI of the movie with the maximum US DVD sales? Is it in the top 25% movies in terms of ROI? Is it in the bottom 25% movies in terms of production budget?

```python
maxDvdSales = df['US DVD Sales'].idxmax()

##get roi
roiSales = df.loc[maxDvdSales, 'ROI']
print(f"ROI with max sales ('{maxDvdSales}'): {roiSales}")

#check roi in top
roiBottom = df['ROI'].quantile(0.75)
if roiSales > roiBottom:
    print(f"The movie '{maxDvdSales}' is in the top 25% of movies.")
else:
    print(f"The movie '{maxDvdSales}' is not in the top 25% of movies.")

#check budget in 25%
budgetTop = df['Production Budget'].quantile(0.25)
budgetMaxSales = df.loc[maxDvdSales, 'Production Budget']
#if else budget
if budgetMaxSales < budgetTop:
    print(f"The movie '{maxDvdSales}' is the bottom 25% of production budget.")
else:
    print(f"The movie '{maxDvdSales}' is not in the bottom 25% of movies.")
```

6. What is the median difference between worldwide gross and US gross? Which movie has the highest difference? (1 point)

```python
#find diff
df['Gross Difference'] = df['Worldwide Gross'] - df['US Gross']

#find median
medianDiff = df['Gross Difference'].median()
print(f"The median is: {medianDiff}")

#highest diff
#use id max for highest diff
highDiff = df['Gross Difference'].idxmax()
#use max for value
highestVal = df['Gross Difference'].max()
print(f"The Movie with the highest difference: '{highDiff}' with a max diff of {highestVal}")
```

7. Do a majority of the movies have IMDB ratings higher than the average? Justify your answer. (2 points)

```python
avgRating = df['IMDB Rating'].mean()

#num of movies above average
count = sum(df['IMDB Rating'] > avgRating)

#half of total of movies
#if count > len(df):
if count > len(df) / 2:
    print(f"Yes. {count} out of {len(df)} movies are above the average rating of {avgRating:.2f}.")
else:
    print(f"No. Only {count} out of {len(df)} movies are above the average rating of {avgRating:.2f}.")
```
