Your code should be structured like this:

def isFibonacci(n):
   
   Write code that checks whether n is a Fibonacci number.
   
   Returns 'Yes' or 'No'

 Print whether 4181 is a Fibonacci number using isFibonacci.

```python
#import math for square root function
import math
#Create new function to do perfect square on new num
def perfSqre(num):
    val = int(math.sqrt(num))
    return val*val == num

def isFibonacci(n):
  #both fib sequeneces are checked and ran giving a True or False value
   return perfSqre(5*n*n + 4) or perfSqre(5*n*n - 4)

#if statement to check if it is true or not so we can print 'Yes' or 'No'
if isFibonacci(4181) == True:
  print("Yes")
else:
  print("No")




```

   Write code that checks whether s1 and s2 are anagrams of each other.
   
   Returns 'Yes' or 'No'

Print whether 'undefinability' and 'unidentifiably' are anagrams using checkAnagrams.

```python
def checkAnagrams(s1, s2):
  #use sorted function to return Yes or No
  #sorted function iterates through the strings
  if(sorted(s1)==sorted(s2)):
    return "Yes"
  else:
    return "No"


if checkAnagrams("undefinability", "unidentifiably") == "Yes":
  print("They ARE anagrams")
else:
  print("They are NOT anagrams")
```
