### Saving plots

There will be cases when we have to use these plots in presentation, articles/reports. In order to that, we need to save these plots so that they can be exported / imported /managed. Different libraries use different methods to save these plots. This NB covers some main methods.

#### Save Matplotlib figures

After creating your plots, here is how you can save matplotlib plots in different formats:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 7, 4.9]
y = [1, 4, 9 ,16, 22, 14]

plt.plot(x,y)
plt.title("X-Y Plot", fontweight = "semibold", fontsize = 14)      # Add plot title
plt.grid(True)                                                     #Add grid background

# These will be saved to your working directory. Add path to save to specific folders.
plt.savefig('plot_name.pdf')      # Save as pdf file

plt.savefig('plot_name.svg')      # Save as svg file

plt.savefig('plot_name.png')      # Save as png file

plt.savefig('plot_name1.png', dpi=300)     # dpi variable is used for quality of png file, increasing quality will also result in increase in image size

plt.savefig('plot_name2.png', dpi=300, transparent=True)   # Export image with transparent background

plt.savefig('plot_name.jpg', dpi=300)   # save file as jpg format. Note jpg files cannot be saved with transparent background

plt.show() # Important: Always use 'plt.show' after saving figures. Otherwise, you will get blank files
```

#### Save seaborn figures

Seaborn figures can be saved using the same method that we used for matplotlib figures. Example

```python
import seaborn as sns
import matplotlib.pyplot as plt


flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")

sns.set(rc = {'figure.figsize':(10,6)})

# Note: you can calculate the relation between different attributes using '.corr()' after your data variable
ax = sns.heatmap(flights.corr(), annot=True, fmt = ".2f")
ax.set_title("Heat Map", fontsize=20)

plt.savefig('seaborn1.pdf')  # save as pdf file

# Every matplotlib export options can be used to save seaborn figures
```
