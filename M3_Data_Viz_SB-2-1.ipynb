### Swarm Plot

<p align="justify">Swarm Plots serve as a form of categorical scatterplots with non-overlapping points, where the points are adjusted(only along categorical axix) in such a way that they don't overlap. This generally scales to better representation of data, but fails to scale well with large number of data values.

````python
import seaborn as sns
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")

tips.head(10)
sns.set(rc={'figure.figsize':(11,8)})
ax = sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
ax.axes.set_title("Swarm Plot", fontsize=20, y=1.01)
````

### Histograms

+ A histogram represents the distribution of a numeric variable.
+ Ordered values in the variable are split into bins, each bin is represented as a bar of the count of the values in it.
+ the width of a bin can (and often needs to be) be adjusted, though the default settings usually work

````python
 #Place data file in working dir. OR enter your own path for file location
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Pokemonstudent.csv',index_col = 0)
data.head()
````

````python
x= data['SpecialAttack']
# sns.displot(x, kind="hist", kde=True, height=6, aspect=1.3).set(title='Basic histrogram')
sns.displot(x, kind="hist", kde=True).set(title='Basic histrogram')

plt.ylabel("Density", labelpad=14)   # Giving label name
plt.xlabel("Special Attack Value", labelpad=14);

# size of plot can be adjusted using 'height' and 'aspect', see below
````

````python
x= data['SpecialAttack']
sns.displot(x, kind="hist", kde=True, height=6, aspect=1.3).set(title='Basic histrogram')

plt.ylabel("Density", labelpad=14)   # Giving label name
plt.xlabel("Special Attack Value", labelpad=14);
````

### Joint Plot

Joint Plots are used to draw relationships between two variables with bivariate and univariate graphs.

````python
data = sns.load_dataset("penguins")

sns.jointplot(data=data, x="bill_length_mm", y="bill_depth_mm", hue="species");
# Method to create joint plot between two variables, categorized by species
````

### Pair Plot   /Scatter plot matrix

Pair Plot is an advanced version of joint plot, where instead of representing relationship between two variables, pair plot represents releationships between all the numerical variables. Example:

````python
# Method to create pairplot, you can also set markers
# Pair plot shows distribution when compared with same variable while show scatter plot for different variables
# You can change the type of distribution and scatter plot as per requiremet
sns.pairplot(data, hue="species", markers=["o", "s", "D"]);
````

### Cat Plot

Cat Plot can be used to represent categorical data. Use ```Catplot``` method to create the cat plot.

````python
data = pd.read_csv('Pokemonstudent.csv',index_col = 0)
````

````python
# Using the catplot function
sns.catplot(data=data, x="Generation", y="HitPoints", color='red', kind = 'point')  # Method to create catplot
plt.title('catplot', fontsize = 14);
````

````python
'''
    'col'      : separate graph by variable
    'col_wrap' : number of graph in row
'''
graph = sns.catplot(data=data, x = 'Type 1', y = 'HitPoints' , kind = 'point', hue = 'Year', col='Generation', col_wrap=3)
graph.set_xticklabels(rotation=90)
````

### Regplot

Regplot method is used to plot data points and apply linear regression model over the data. More information about linear regression is available [here](https://en.wikipedia.org/wiki/Linear_regression). Let's see an example of using Regplot:

````python
'''
Using regplot plot along with facetgrid

    col : divide graphs by variable
    col_wrap : number of graph in row
    marker : change marker type
'''

graph = sns.FacetGrid(data, col='Generation', col_wrap=3)
graph.map(sns.regplot, 'SpecialAttack', 'SpecialDefense', color='green', marker='+')
````

### KDE Plot

The kernel density estimate  (KDE) plot can be used for a range of insughts including visualizing univariate or bivariate distributions. KDE represents the data using a continuous probability density curve in one or more dimensions.

````python
data = sns.load_dataset("penguins")
````

````python
g = sns.FacetGrid(data, col="species", hue='species')
g.map(sns.kdeplot, "bill_length_mm", "bill_depth_mm", fill=True)
````

### Pair Plot   / Scatter plot matrix

Pair Plots can be viewed as being an advanced version of joint plots, where instead of representing the relationship between two variables, pair plots represents releationships between all the numerical variables, and can account for factor levels.

````python
sns.pairplot(data, hue="species", markers=["o", "s", "D"], diag_kind="hist"); # Use 'diag_kind' for plotting histograms
````

We can also use different graphs type for different parts of the grid

````python
graph = sns.pairplot(data, hue="species", markers=["o", "s", "D"], diag_kind="hist")
graph.map_lower(sns.kdeplot)
graph.map_upper(sns.regplot);
````
