Question 1

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

s1 = pd.Series(np.random.randint(10, 20, 10))
s2 = pd.Series(np.random.randint(30, 40, 10))

Product = pd.DataFrame({'Price': s1, 'Quantity': s2})
print(Product)
```

Question 2

```python
sortProduct = Product.sort_values(by='Price', ascending=False)
print(sortProduct)
```

Question 3

```python
avgPrice = Product['Price'].mean()

print(avgPrice)
```

Question 4

```python
Product['Total cost'] = Product['Price'] * Product['Quantity']
print(Product)
```

Question 5

```python
maxValue = Product['Total cost'].max()
maxValue
```

Question 6

```python
findMax = Product['Total cost'].idxmax()
#use .loc()
rowMax = Product.loc[findMax]
print(rowMax)
```
