Each of the following are plots that are used in the research paper

First we showcase the data set and see what is in it

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

data = pd.read_csv('/content/sample_data/TwitterDataWithSentiment_v2.csv')
data

```

```python
#Scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['followers_count'], y=data['listed_count'])
plt.title("Follower count vs Listed count")
plt.show()
```

```python
#heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
#boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['protected'], y=data['favourites_count'])
plt.title("Protected or not vs favorites count")
plt.show()
```

```python
#word cloud
text_data = " ".join(text for text in data['text'])
wordcloud = WordCloud(background_color="white").generate(text_data)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud")
plt.show()
```

```python
#violin
plt.figure(figsize=(10, 6))
sns.violinplot(x=data['source'], y=data['followers_count'])
plt.title("source vs followers count")
plt.show()
```
