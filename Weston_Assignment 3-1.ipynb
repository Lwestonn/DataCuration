5. We have used the file TwitterData.csv in our week 3 material to generate WordClouds and Visualizations. Using the same data:

* Read the first 280 rows of the CSV file into a Data Frame.
* Use the STOPWORDS provided by NLTK, identify atleast 3 more stopwords in the dataset and add it to the stopword list.
* Generate a WordCloud of unigrams with a White Background.   

```python
import nltk
import pandas as pd
import ssl
import string
import matplotlib.pyplot as plt
from wordcloud import WordCloud

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

nltk.download(["stopwords", "vader_lexicon"])

from nltk.corpus import stopwords



data = pd.read_csv("/content/sample_data/TwitterDataWithSentiment.csv", nrows = 280)

#print(data)

stopwords_set = set(stopwords.words('english'))

additional_stopwords = ['like', 'see', 'got']

stopwords_updated = stopwords_set.union(additional_stopwords)


def remove_stopwords(text):
    filtered_words = [word.lower() for word in text.split() if word.lower() not in stopwords_updated]
    return " ".join(filtered_words)

data['text'] = data['text'].apply(remove_stopwords)

#print(data)


combined_text = " ".join(row for row in data['text'])

#create wordcloud
wordcloud = WordCloud(background_color="white").generate(combined_text)

#display wordcloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
