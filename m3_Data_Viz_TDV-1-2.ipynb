# Working with Text Data


* Let us use a sample dataset to perform analysis on it.

* Before performing the data visualization, we need to clean the data.

* For example, let's say our text data has a sentence, "Hi, I am a freshman at Rutgers University and I wish to be a Data Scientist."

* The words like "Hi", "I", "am", "to", "be" are often treated as 'stop words'.

* A stop word is a most common word in a language and often dont add much value in terms of information to the sentence.

* Reason to remove: they are too common, occur freuently, and can often clutter the visualization space without providing additional insights

* Since these words are of almost no importance in analysing the text, we often remove them.

* To remove them, Python has a module named nltk. The nltk word corpus consists of the set of stop words.

* Let's import the data, understand the data and try to clean it.



````python
import pandas as pd
import string

# Opening the data file
# f = open('./drive/MyDrive/data_1.txt','r')       #Colab
f = open('m3_data_1.txt','r')

# Reading the data into a variable.
data = f.readlines()
````

* Based on the above, we have created (imported m3_data_1.txt) data as f.

* To read the actual data, we use the methods like read(), readlines() or readline().

* Each method has a different functionality as follows:
> * read(): This method reads the text file into a single string.
> * readlines(): This method reads the text file into a list where every element in the list is a line in the text file.
> * readline(): This method is used to read the text file line by line. Since we can't determine the number of lines beforehand, we use a loop to iterate through every line and read it.

* Now, lets print the data and see how the data is read.

````python
data
````

* In the above data, the first element is the first line of the text file, the second element is the second line and so on.
* Therefore, the size of this array is equal to the number of lines in the text file.
* An observation that can be seen on the above array is that the last characters in every line is "\n".
* This is a special character which signifies a new line.
* Now, lets use the nltk module to remove the stop words.

````python
import nltk

import ssl

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

nltk.download(["stopwords", "vader_lexicon"])

from nltk.corpus import stopwords


````

* We use the ssl module to download the data for nltk module

* The package nltk contains different modules within itself and the line nltk.download() downloads individual module of the package.

* Now, lets print the stopwords present in the english language.

````python
print(stopwords.words('english'))
````

* Now, lets write a function to remove stop words from the sentences.

* We iterate through every word and if the word is present in the stopwords list, we don't add them to our text list.

````python
# Creating a set of stopwords
stopwords_set = set(stopwords.words('english'))

# This function iterates over every word. If the word is present in the
def remove_stopwords(text):
    filtered_words = [word.lower() for word in text.split() if word.lower() not in stopwords_set]
    return " ".join(filtered_words)

print(data)
for i in range(len(data)):
    data[i] = remove_stopwords(data[i])

print(data)


````

* From the above print statements, we can observe the difference between the list before removing the stop words vs after removing the stop words.

* Note that every word is converted into the lower case because of the **lower()** method in the function.

* We will now create the corpus of the text data. Using this corpus, we can create multiple data visualizations like word cloud andpacked bubble charts.

* This corpus consists a dictionary with key as the word and the value as the frequency of the word.

* Just like a hashmap in other languages like Java and C++, a dictionary is a key-value pair data structure.



````python
corpus = {}
for sentence in data:
  for word in sentence.split(" "):
    if word not in corpus:
      corpus[word] = 1
    else:
      corpus[word] += 1

print(corpus)
````

* The above corpus shows the words along with it's frequencies in the data.
* Now let's plot the frequency versus word plot. A basic visualization that is done on the textual data.

````python
# Here, we sort the corpus in the descending order
corpus = sorted(corpus.items(), key = lambda x: x[1], reverse=True)

# Getting the first 15 items from the corpus
corpus = corpus[:15]

import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (10, 5))

xvalues = []
yvalues = []

for key, value in corpus:
  xvalues.append(key)
  yvalues.append(value)

ax.bar(xvalues, yvalues, edgecolor="black");
ax.set_xticklabels(xvalues, rotation=90, va='top')

plt.title('Frequency Distribution')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
````

* From the above plot, most of the words occur once, there is one word with the frequency as 4, some with 3 and and most with 2.

* The words on the x-axis will be unclear due to the length of the data - if too many words are plotted.  

* However, this can be made more clear by using an interactive plot which clearly highlights the word we are pointing on.

# WordClouds

* Wordclouds, also known as Tag Clouds, is a data visualization technique that is used for representing textual data.

* Visually, a word cloud looks like a group of words where some words have bigger font and some of them have a smaller font.

* The words with the bigger font size signifies that the occurrence (frequency or count) of the word is high and could therefore be more important.

* Therefore, the words with higher font size are more likely to dsiplay the main themes in the text.

* There are many ways to determine the importance of the word. Usually, the frequency of the word is taken into consideration.

* This means that the word/s that has/have occurred the most number of times could represent a dominant insight/s.

* Identifying high frequency words are often useful, BUT by itself this should not be used to conclude on main themes - additional in-depth analysis using multiple methods is necessary.



# Applications of Word Clouds

Word clouds are used in many scenarios. Some of the most popular use cases are as follows:

1. **Customer's Feedback:**
> * Wordclouds are widely used by the stakeholders to analyze the feedback received.
> * Say a company like Amazon wants to analyze the feedback it recived on the product with say about 10,000 reviews.
> * It would be very difficult for them to read through every review and make a note of it. Instead, a usual approach which they follow is Word Cloud.
> * They create word clouds for reviews, often by category such as male /female or age groups, and analyze it to understand the tone, the words frequently used, the sentiment of the feedback, etc;
> * Based on these reviews, it would be easier for the firm to take future decisions.

2. **Trending Topics:**
> * This method is also used to find trending topics.
> * For example, analyzing tweets and finding the most tweeted word could give us insights into trends /viral topics.

3. **... and a range of other purposes in presentations, reports and research studies across domains such as politics, media, stock market and marketing**

# Creating a WordCloud in a Specific Colour Pattern

* Imagine we have a mask and we need to retain the background of the mask.
* Let's take a sample mask, an Indian flag for this example.
* Just like the twitter, lets download the Indian Flag from this url.

````python
import urllib.request
imgURL = "https://upload.wikimedia.org/wikipedia/commons/b/bc/Flag_of_India.png"

urllib.request.urlretrieve(imgURL, "flag.png")
````

````python
from imageio import imread
flag_mask = imread('flag.png')
````

````python
wc_india = WordCloud(collocations=False,stopwords=stopwords, background_color="white", mode="RGBA", max_words=1000, mask=flag_mask).generate(data)

image_colors = ImageColorGenerator(flag_mask)
plt.figure(figsize=[10,20])
plt.imshow(wc_india.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")

plt.title('Word Cloud Title')

# store to file
plt.savefig("flag_cloud.png", format="png")

plt.show()
````

## Using a single colour for the word cloud

* Inorder to use a single color to the word cloud, we pass the color_func parameter to the imshow.
* Note that in the above example, we set this parameter to the **image_colors** function.
* Therefore, it clearly explains the purpose of **ImageColorGenerator** class in the above example.
* It creates the colours of the image such that the text within the image is the same colour as the background colour.  

````python
wc_india = WordCloud(collocations=False,stopwords=stopwords, background_color="white", mode="RGBA", max_words=1000, mask=flag_mask).generate(data)

plt.figure(figsize=[10,20])
plt.imshow(wc_india.recolor(color_func=lambda *args, **kwargs: "black"), interpolation="bilinear")
plt.axis("off")

plt.title('Word Cloud Title')

# store to file
plt.savefig("flag_cloud.png", format="png")

plt.show()
````

* If we wish to consider both the singular words and plural words as a two different words, then we pass the normalize_plurals as False.
* By default, its value is True.

````python
wc_india = WordCloud(collocations=False,stopwords=stopwords, background_color="white", mode="RGBA", normalize_plurals=False, max_words=1000, mask=flag_mask).generate(data)

plt.figure(figsize=[10,20])
plt.imshow(wc_india.recolor(color_func=lambda *args, **kwargs: "black"), interpolation="bilinear")
plt.axis("off")

plt.title('Word Cloud Title')

# store to file
plt.savefig("flag_cloud.png", format="png")

plt.show()
````

* You can see that when we pass the normalize_plurals to False, the occurrences of singular and plural words exist separately.
* In the above case, gets and says are such words which don't exist in the above example and exist in this example.
