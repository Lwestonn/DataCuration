Part 1 LUCAS WESTON

```python
book1 = {
    'id': 1,
    'title': '1984',
    'author': 'George Orwell'
}

book2 = {
    'id': 2,
    'title': 'War and Peace',
    'author': 'Leo Tolstoy'
}

book3 = {
    'id': 3,
    'title': 'The Shining',
    'author': 'Stephen King'
}


books = [book1, book2, book3]
```

Part 2

```python
def add_book(book_id, title, author):

    for book in books:
        if book['id'] == book_id:
            print(f"That book ID already exists{book_id}")
            return

    new_book = {
        'id': book_id,
        'title': title,
        'author': author
    }

    books.append(new_book)
```

Part 3

```python
add_book(4, 'To Kill a MockingBird', 'Harper Lee')
```
